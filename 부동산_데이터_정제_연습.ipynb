```python
#공백 제거
df['분양가격(㎡)'] = df['분양가격(㎡)'].str.strip()
```

```python
df.loc[df['분양가격(㎡)']==' ']
```

```python
df.loc[df['분양가격(㎡)'] == '', '분양가격(㎡)']=0 #분양가격 ''행에 0 넣어주기
df['분양가격(㎡)'] = df['분양가격(㎡)'].fillna(0) #Nan값에도 0 넣기
```

```python
df.loc[df['분양가격(㎡)']=='6,657']
df['분양가격(㎡)'] = df['분양가격(㎡)'].fillna(0)
```

```python
df['분양가격(㎡)'] = df['분양가격(㎡)'].str.replace(',','') #콤마 위치 찾아 콤마도 제거
df.iloc[2125]
```

```python
df['분양가격(㎡)'] = df['분양가격(㎡)'].str.replace('-','')
df['분양가격(㎡)'] = df['분양가격(㎡)'].fillna(0)
df.loc[df['분양가격(㎡)'] == ''] = 0
df['분양가격(㎡)'] = df['분양가격(㎡)'].astype(int)
```

```python
#'규모구분' column에 불필요한 '전용면적 제거하기
df['규모구분'].value_counts()
```

```python
df['규모구분'] = df['규모구분'].str.replace('전용면적','')
df['규모구분'].value_counts()
```
